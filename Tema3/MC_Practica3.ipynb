### **<font color="Navy"> Ejercicio 3: La pena climática  (12 puntos) </font>**

La "pena climática" es un término que se refiere al resultado de la reducción de la emisión de aerosoles, los cuales pueden tener efectos tanto en la calidad del aire como en el clima. Los aerosoles atmosféricos tienen, en promedio, un efecto de enfriamiento en el clima al reflejar la radiación solar de vuelta al espacio, contrarrestando parcialmente el calentamiento causado por los gases de efecto invernadero. Por otro lado, algunos aerosoles pueden ser contaminantes del aire y contribuir a la mala calidad del aire y a problemas de salud pública.

Cuando se reduce la emisión de aerosoles como resultado de políticas ambientales que buscan mejorar la calidad del aire, como la reducción de la quema de combustibles fósiles, se podría producir un fenómeno conocido como "pena climática". Esta "pena", o castigo sería una traducción más adecuada, se refiere al hecho de que la disminución de los aerosoles puede eliminar parte de su efecto de enfriamiento en el clima, lo que lleva a un aumento neto de la temperatura global.

### **<font color="Navy"> Ejercicio 7: El cambio en el nivel del mar  (18 puntos) </font>**

Kopp et al., (2016), https://www.pnas.org/doi/epdf/10.1073/pnas.1517056113, escribieron una parametrización sencilla para relacionar el cambio de temperatura global con el cambio en el nivel del mar ($h$). Dicha parametrización ha sido utilizada en FaIR para estimar los cambios en el nivel del mar esperados dados los resultados de FaIR. La parametrización completa se puede escribir a partir del sistema de ecuaciones diferenciales:

\begin{align}
\frac{dh}{dt}=a(T(t)-T_0(t))+c(t) \,\,\,\ \textbf{(1)}
\end{align}

donde $a$ es la sensitividad del cambio de nivel del mar a una desviación de $T(t)$ del equilibrio $T_0(t)$. Esta ecuación diferencial están acoplada a las ecuaciones:

\begin{align} 
\frac{dT_0(t)}{dt}=\frac{(T-T_0(t))}{\tau_1} \,\,\,\ \textbf{(2)}
\end{align}

\begin{align}
\frac{dc(t)}{dt}=\frac{\textbf{c}}{\tau_2}   \,\,\,\ \textbf{(3)}
\end{align}

donde $\tau_1$ es la escala de tiempo con la que la temperatura se relaja hacia la temperatura de equilibrio y $\textbf{c}$ es una tasa de cambio con un decaimiento exponencial dado por $\tau_2$. Este conjunto de ecuaciones representan un modelo simple de cambio en el nivel promedio del mar como resultado de un aumento de temperatura. Específicamente, este modelo utiliza un valor de equilibrio de $T_0$ representativo del siglo XX, y los desarrolladores de FaIR 

En FaIR podemos usar el parámetro $a$ con un valor de 5, $\tau_1=102$,  $\tau_2=3392$ y el parámetro $\textbf{c}$ como 0.03.
